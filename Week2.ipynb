```python
# Implement a binary search function
def binarySearch(array, element, start, end):
    if start > end:
        return -1

    mid = (start + end) // 2
    if element == array[mid]:
        return mid

    if element < array[mid]:
        return binarySearch(array, element, start, mid-1)
    else:
        return binarySearch(array, element, mid+1, end)


array=[1,2,3,5,6,8,10,12,15,19,30,45]
print(binarySearch(array,5,2,19))
print(binarySearch(array,2,1,15))
```

```python
##findItemInSortedListSlow--one of search functions

def findItemInSortedListSlow(myList, item):
    i=0
    while i<len(myList):
        if myList[i]==item:
            return (True,i+1)
        i = i+1
    return (False,i)  
```

```python
##findItemInSortedListFast--one of search functions


def findItemInSortedListFast(myList, item):
    lowerLimit = 0 # smallest possible index for the item
    upperLimit = len(myList)-1 # largest possible index for the item
    i=0 
    while lowerLimit <= upperLimit:
        i=i+1
        # check if the item is at midpoint. If it is, return True. If it is not, halve the range to the left half if 
        # the item at midpoint is bigger that the item we are searching for and the right otherwise
        midpoint = (upperLimit + lowerLimit) // 2
        if myList[midpoint]==item:
            return (True,i)
        elif myList[midpoint]<item:
            lowerLimit = midpoint+1
        else:
            upperLimit = midpoint-1
    return (False,i+1)
```

```python
#the plot of the running time for findItemInSortedListFast

randomlistfortime = list() 
timeforfindfast = list() 
for i in range(1, 10): 
    #gernerate random numnbers
    randomnumber = randint(0, 100 * i, 100 * i)
    #calculate the running time 
    start_time_findfast = time.perf_counter() 
    findItemInSortedListFast(randomnumber,5)
    end_time_findfast = time.perf_counter()
    print("The runnung time for findItemInSortedListFast is ", end_time_findfast - start_time_findfast) 
    randomlistfortime.append(len(randomnumber)) 
    timeforfindfast.append(end_time_findfast - start_time_findfast) 

plt.xlabel('List') 
plt.ylabel('Running Time') 
plt.plot(randomlistfortime , timeforbinarysearch, label ='findItemInSortedListFast')
plt.show() 
```

```python
# Implemnet a insertion sort
 
def insertionSort(array): 
 
    for i in range(1, len(array)): 
        key = array[i] 
        j = i-1
        #move the items in the list to inset the i directly
        while j >= 0 and key < array[j] : 
                array[j + 1] = array[j] 
                j -= 1
        array[j + 1] = key  
```

```python
#the plot of running time for insertion sort

randomlistfortime = list() 
timeforinsertsort = list() 
for i in range(1, 10): 
    #gernerate random numnbers
    randomnumber = randint(0, 100 * i, 100 * i)
    #calculate the running time
    start_time_insertsort = time.perf_counter() 
    insertionSort(randomnumber)
    end_time_insertsort = time.perf_counter()
    print("The runnung time for insertion sort is ", end_time_insertsort - start_time_insertsort) 
    randomlistfortime.append(len(randomnumber)) 
    timeforinsertsort.append(end_time_insertsort - start_time_insertsort) 

plt.xlabel('List') 
plt.ylabel('Running Time') 
plt.plot(randomlistfortime , timeforinsertsort, label ='insertionsort') 
plt.show() 
```

```python
# Implement a merge sort
def mergeSort(array):
    if len(array) > 1:
        mid = len(array) // 2
        left = array[:mid]
        right = array[mid:]

        mergeSort(left)
        mergeSort(right)
        
        i = 0
        j = 0
        k = 0
        
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                array[k] = left[i]
                i += 1
            else:
                array[k] = right[j]
                j += 1
            k += 1
            
        while i < len(left):
            array[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            array[k]=right[j]
            j += 1
            k += 1           
```

```python
#the plot of running time for merge sort
randomlistfortime = list() 
timeformergesort = list() 
for i in range(1, 10):
    #gernerate random numnbers
    randomnumber = randint(0, 100 * i, 100 * i)
    #calculate the running time
    start_time_mergesort = time.perf_counter() 
    mergeSort(randomnumber)
    end_time_mergesort = time.perf_counter()
    print("The runnung time for merge sort is ", end_time_mergesort - start_time_mergesort) 
    randomlistfortime.append(len(randomnumber)) 
    timeformergesort.append(end_time_mergesort - start_time_mergesort) 

plt.xlabel('List') 
plt.ylabel('Running Time') 
plt.plot(randomlistfortime , timeformergesort, label ='mergesort') 
plt.show()
```

For the plots as shown above, we found that merger sort is more efficient than insertion sort.
